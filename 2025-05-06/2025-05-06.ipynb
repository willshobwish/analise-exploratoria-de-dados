```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from scipy.stats import shapiro
```

```python
df = pd.DataFrame({"x":[7.21,7.31,7.56,8.08,8.22,8.3,8.61],"y":[15.76,16.66,17.24,18.66,18.92,19.29,19.86]})
```

```python
plt.plot(df["x"],df["y"],'o')
plt.xticks(np.linspace(df["x"].min(),df["x"].max(),10))
plt.yticks(np.linspace(df["y"].min(),df["y"].max(),10))
plt.grid()
plt.xlabel("Massa corporal sem gordura")
plt.ylabel("Taxa de metabolismo")
```

```python
lr = LinearRegression()
lr.fit(np.array(df["x"]).reshape(-1, 1),df["y"].tolist())
predictions = lr.predict(np.array(df["x"]).reshape(-1, 1))
plt.xticks(np.linspace(df["x"].min(),df["x"].max(),10))
plt.yticks(np.linspace(df["y"].min(),df["y"].max(),10))
plt.grid()
plt.xlabel("Massa corporal sem gordura")
plt.ylabel("Taxa de metabolismo")
plt.plot(df["x"],df["y"],'o')
plt.plot(df["x"],predictions)
for x, y in zip(df["x"],predictions):
    print(x,y)
```
